```python
palavras = """Willkommen
Benvinguda
Huānyíng
Bienvenido
Vitajte
Bienvenue
Maligayang
Tervetuloa
Fogadtatás
Welkom
Welcome
Benvenuto
Fáilte
Selamat datang
Naumai
Velkommen
желанный
Välkommen
Vítejte""".split("\n") * 2
```

```python
import random
from itertools import chain
w = "Bem-vindos"
text = { p: random.randint(1,5) * .1 for p in palavras}
text[w] = 2
```

```python
from wordcloud import WordCloud
width = 918
height = 120

wordcloud = WordCloud(collocations=False,
                      font_path="/usr/local/lib/python3.7/dist-packages/matplotlib/mpl-data/fonts/ttf/BowlbyOneSC-Regular.ttf?raw=true",
                      background_color="white",
                      width=width,
                      height=height,
                      repeat=True,
                      colormap="prism",

                      relative_scaling = 0.2).generate_from_frequencies(text)

# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
plt.subplots(1, 1, figsize=(width // 30, height // 30), dpi=300)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off");
```
